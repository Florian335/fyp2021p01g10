# loading the data

```python
# loading the files
file = "C:/Users/Ruben PC/Desktop/FirstYearProject/1st_project/"
accidents = pd.read_csv("data/processed/Road Safety Data - Accidents 2019.csv") 
vehicle = pd.read_csv("data/processed/Road Safety Data - Vehicles 2019.csv")
casualty = pd.read_csv("data/processed/Road Safety Data - Casualties 2019.csv")



# combining the data into one combined dataFrame
result = pd.concat([accidents, vehicle, casualty], axis=1)

# removing duplicate columns
data = result.loc[:,~result.columns.duplicated()]
sheffield = (result["Local_Authority_(District)"] == 215)
data = data[sheffield]
```

# Five Number Summary, Boxplot

```python
#NUMBER OF VEHICLES IN AN ACCIDENT
mask_num_vehicle = (accidents['Number_of_Vehicles']>0)
NumVehicles = accidents["Number_of_Vehicles"][mask_num_vehicle]
NumVehicles = NumVehicles.to_frame()
NumVehicles.describe(), NumVehicles.boxplot()
```

```python
#AGE OF VEHICLE
mask_age_vehicle = (vehicle['Age_of_Vehicle']>=0)
AgeVehicles = vehicle["Age_of_Vehicle"][mask_age_vehicle]
AgeVehicles = AgeVehicles.to_frame()
AgeVehicles.describe(), AgeVehicles.boxplot()
```

```python
#AGE OF DRIVER
mask_age = (vehicle['Age_of_Driver']>=18)
Age_Driver = vehicle["Age_of_Driver"][mask_age]
Age_Driver = Age_Driver.to_frame()
Age_Driver.describe(), Age_Driver.boxplot()
```

```python
#AGE OF MALE DRIVERS
mask_male = (vehicle['Sex_of_Driver']==1)
Age_MaleDrivers = vehicle["Age_of_Driver"][mask_age][mask_male]
Age_MaleDrivers = Age_MaleDrivers.to_frame()
Age_MaleDrivers.describe(), Age_MaleDrivers.boxplot()
```

```python
#AGE OF FEMALE DRIVERS
mask_female = (vehicle['Sex_of_Driver']==2)
Age_FemaleDrivers = vehicle["Age_of_Driver"][mask_age][mask_female]
Age_FemaleDrivers = Age_FemaleDrivers.to_frame()
Age_FemaleDrivers.describe(), Age_FemaleDrivers.boxplot()
```

```python
#AGE Of CASUALTY
mask_age2 = (casualty['Age_of_Casualty']>=0)
AgeCasualty = casualty["Age_of_Casualty"][mask_age2]
AgeCasualty = AgeCasualty.to_frame()

AgeCasualty.describe(), AgeCasualty.boxplot(),
```
